## Задание 1


    Треугольником Паскаля называется числовой треугольник, в котором по краям стоят единицы, а каждое число внутри равно сумме двух стоящих над ним сверху значений. 
    Дано натуральное число n. Получить первые n строк треугольника Паскаля.

```python
n=int(input()) # ввод строк
x=[1,1] #2 строка
x1=[1] #1 строка
r=0 # обычный счетчик. проходит от 2 строки до n
p=" " # Будет вставлять пробелы чтобы был треугольник
for y in range(n):
    p+=" " # делает первую строку с наибольшим числом пробелов
if n>0: # если больше 1 строки
    print(p,1)
p = p.replace(' ', '', 1) 
if n>1: # если больше 2 строк
    print(p,1,1)
elif n<1: # если меньше 1 сторки
    print("Как я сделаю треугольник ааааа?")
while n>2: # если больше 2 строк
    while r!=len(x)-1:
        x1.append(x[r]+x[r+1]) # добавит сумму верхней строки в следующий индекс
        r+=1 # пройдем дальше по строке чтобы ссумировать все соседние числа
    x1.append(1) # в конце всегда 1 поэтому добавим в конце строки 1
    p = p.replace(' ', '', 1) # уберет 1 пробел из строки p
    print(p,*x1)
    n-=1 # убираем одну строку
    x=x1 # x будет предыдущая строка
    x1=[1] # следующая строка начинающаяся с 1
    r=0 # обнуляем счётчик
```

## Задание 3

    Пользователь вводит размер n квадратной матрицы A и её элементы. . Написать программу, которая выводит обратную матрицу (т.е A^-1) на экран. Для этого реализуйте один из методов: метод алгебраических дополнений или метод исключения неизвестных Гаусса. 

```python
import numpy as np
n=int(input('Напишите размер матрицы:'))
x=0 # счётчик
mas1=[]
mas=[0]*n # заполняем матрицу пустыми элементами
for i in range(n): 
    mas[i]=[0]*n 
while x<n: # запрашиваем строки и вписываем в каждый столбик
    print('Напишите',x+1,'строку')
    mas1=list(map(int, input().split(","))) # Ввод строк, они будут проиндексированны
    mas[x]=mas1 # индекс начинается с 0 и это будет верхняя строка 
    x+=1
result=np.linalg.inv(mas) # готовую матрицу делаем обратной
print(result)
```

## Задание 4

Даны целые числа 0,1,2,3,...,63. Получить квадратную матрицу (таблицу) 8 на 8, элементами которой являются указанные числа, расположенные в ней по схемам, приведённым ниже.

```python
mas=[]
for x in range(64):
    mas.append(x) # сделали массив из 64 элементов
mas=np.reshape(mas, (-1, 8)) # преобразовали в матрицу 8 на 8, -1 показывает что мы ушли из плоскости
for x in range(8): # мы немного схалявили,но по условию всё сделано правильно
    mas[x][0]=x
    mas[x][2]=16+x
    mas[x][4]=32+x
    mas[x][6]=48+x
    mas[x][1]=15-x
    mas[x][3]=31-x
    mas[x][5]=47-x
    mas[x][7]=63-x
print(mas)
```
